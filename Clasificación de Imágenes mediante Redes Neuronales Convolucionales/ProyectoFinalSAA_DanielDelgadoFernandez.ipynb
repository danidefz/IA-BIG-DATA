## 4. MODELO 1: CNN DESDE CERO

```python
print("\n--- 4. Modelo 1: CNN desde Cero ---")

# CELDA 4.1: Construcción de la Arquitectura
def build_cnn_from_scratch(image_shape, angle_shape, num_classes=1):
    # Entrada para las imágenes
    image_input = Input(shape=image_shape, name='image_input')
    # Entrada para el ángulo de incidencia
    angle_input = Input(shape=angle_shape, name='angle_input')

    
    x = Rescaling(1./255)(image_input)

    x = Conv2D(32, (3, 3), activation='relu', padding='same')(image_input) 
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)
    
    # Aplanar la salida de la CNN
    image_features = Flatten()(x)

    # Concatenar las características de la imagen con el ángulo de incidencia
    concatenated_features = Concatenate()([image_features, angle_input])

    # Capas densas para clasificación
    z = Dense(256, activation='relu')(concatenated_features)
    z = BatchNormalization()(z)
    z = Dropout(0.5)(z)
    
    # Como es clasificación binaria (iceberg o no), usamos 1 neurona con activación sigmoide
    output = Dense(num_classes, activation='sigmoid', name='output')(z)

    # Crear el modelo
    model = Model(inputs=[image_input, angle_input], outputs=output, name="cnn_from_scratch")
    return model


```

#### 4.1. Construcción de la Arquitectura

```python
# --- TU CÓDIGO AQUÍ ---
# Define tu modelo CNN secuencial. Asegúrate que la capa de entrada
# coincida con IMAGE_SIZE y IMAGE_CHANNELS. La salida debe tener NUM_CLASSES neuronas.
# Crear el modelo
img_shape_cnn = (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS)
angle_shape_cnn = (1,) # El ángulo es un solo valor
model_scratch = build_cnn_from_scratch(img_shape_cnn, angle_shape_cnn)
model_scratch.summary()

# --- FIN TU CÓDIGO ---
```

#### 4.2. Compilación del Modelo

```python
# --- TU CÓDIGO AQUÍ ---
# Elige un optimizador, una función de pérdida y métricas

LEARNING_RATE = 0.001 # Puedes ajustar esto
optimizer = tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE)

model_scratch.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)
print("Modelo compilado.")


# Callbacks
early_stopping = EarlyStopping(
    monitor='val_auc', 
    patience=10,       
    verbose=1,
    mode='max',        
    restore_best_weights=True 
)
# --- FIN TU CÓDIGO ---
```

#### 5.1. Cálculo de Pesos de Clase (si hay desbalanceo)

```python
# --- TU CÓDIGO AQUÍ ---
# Si el EDA mostró desbalanceo, calcula los pesos
from sklearn.utils import class_weight

# Calcular pesos de clase basados en y_train (asegúrate que y_train esté definido de la sección 3)
# y_train contiene las etiquetas (0 o 1)
if 'y_train' in globals():
    class_weights_values = class_weight.compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weights_dict = dict(enumerate(class_weights_values))
    print(f"Pesos de clase calculados: {class_weights_dict}")

    # Verificar si hay un desbalanceo significativo para decidir si usar class_weight
    # Por ejemplo, si una clase tiene el doble de peso que la otra.
    # Esta es una heurística simple, puedes ajustarla.
    use_class_weighting = True # O una lógica basada en la diferencia de pesos
    if not use_class_weighting:
        class_weights_dict = None # No usar pesos si no es necesario
        print("Desbalanceo no considerado suficientemente significativo para aplicar class_weighting.")
else:
    print("Variable y_train no encontrada. Asegúrate de haber ejecutado la división de datos.")
    class_weights_dict = None
    use_class_weighting = False
# --- FIN TU CÓDIGO ---
```

#### 5.2. Definición de Data Augmentation

```python
# Se puede hacer con capas de Keras o con ImageDataGenerator
# --- TU CÓDIGO AQUÍ ---
# Capas de Data Augmentation para las imágenes
data_augmentation_layers = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"), # Adecuado para imágenes de radar/satélite
    layers.RandomRotation(0.2), # Rotaciones leves
    # layers.RandomZoom(0.1), # Zoom leve si aplica a este tipo de imágenes
    # layers.RandomContrast(0.1), # Ajuste leve de contraste si aplica
], name="data_augmentation")

# Nota: El data augmentation se aplicará selectivamente a la rama de imágenes del modelo.
# --- FIN TU CÓDIGO ---
```

#### 5.3. Construcción y Entrenamiento del Modelo Mejorado

```python
# --- TU CÓDIGO AQUÍ ---
# Función para construir el modelo CNN mejorado con Data Augmentation
# Asumimos que build_cnn_from_scratch está definida como en la sección 4.
# Vamos a modificarla o crear una nueva que incorpore la capa de augmentation.

def build_cnn_augmented(image_shape, angle_shape, data_augmentation_model, num_classes=1):
    # Entrada para las imágenes
    image_input = Input(shape=image_shape, name='image_input_aug')
    # Entrada para el ángulo de incidencia
    angle_input = Input(shape=angle_shape, name='angle_input_aug')

    # Aplicar Data Augmentation a la entrada de imágenes
    augmented_images = data_augmentation_model(image_input)

    # Rama CNN para procesar imágenes aumentadas
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(augmented_images)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)
    
    image_features = Flatten()(x)

    # Concatenar las características de la imagen con el ángulo de incidencia
    concatenated_features = Concatenate()([image_features, angle_input])

    # Capas densas para clasificación
    z = Dense(256, activation='relu')(concatenated_features)
    z = BatchNormalization()(z)
    z = Dropout(0.5)(z)
    
    output = Dense(num_classes, activation='sigmoid', name='output_aug')(z)

    model = Model(inputs=[image_input, angle_input], outputs=output, name="cnn_augmented")
    return model

# Crear el modelo mejorado
# IMAGE_CHANNELS, IMAGE_HEIGHT, IMAGE_WIDTH, SEED, BATCH_SIZE, EPOCHS_SCRATCH deberían estar definidos
img_shape_aug = (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS)
angle_shape_aug = (1,)
model_augmented = build_cnn_augmented(img_shape_aug, angle_shape_aug, data_augmentation_layers)

# Compilar el modelo mejorado
# LEARNING_RATE ya debería estar definido
optimizer_aug = tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE) # Puedes usar el mismo o uno diferente

model_augmented.compile(
    optimizer=optimizer_aug,
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)
model_augmented.summary()
print("Modelo mejorado compilado.")

# Callbacks (EarlyStopping ya debería estar definido)
# early_stopping ya está definido
# model_checkpoint_path_augmented = os.path.join(MODEL_FOLDER, 'model_augmented_best.keras')
# model_checkpoint_augmented = ModelCheckpoint(
#     filepath=model_checkpoint_path_augmented,
#     monitor='val_auc',
#     save_best_only=True,
#     mode='max',
#     verbose=1
# )

print(f"\nIniciando entrenamiento del modelo mejorado por {EPOCHS_SCRATCH} épocas...")
# Asegúrate que train_data_model, y_train, val_data_model, y_val estén disponibles
# train_data_model = [X_img_train, X_ang_train]
# val_data_model = [X_img_val, X_ang_val]

history_augmented = model_augmented.fit(
    train_data_model, 
    y_train,
    epochs=EPOCHS_SCRATCH, # Puedes usar EPOCHS_SCRATCH o definir EPOCHS_AUGMENTED
    batch_size=BATCH_SIZE,
    validation_data=(val_data_model, y_val),
    callbacks=[early_stopping], # Añadir model_checkpoint_augmented si se usa
    class_weight=class_weights_dict, # Aplicar pesos de clase si se definieron
    verbose=1
)
# Puedes reutilizar la arquitectura de `model_scratch` o definir una nueva.
# Si usas capas de augmentation, añádelas al principio.
# --- FIN TU CÓDIGO ---
```

## 6. MODELO 3: TRANSFER LEARNING

```python
print("\n--- 6. Modelo 3: Transfer Learning ---")
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess
```

### 6.1. Selección y Carga del Modelo Base Pre-entrenado

```python
# --- TU CÓDIGO AQUÍ ---
# Elige una red base (VGG16, ResNet50V2, MobileNetV2, etc.)
# Asegúrate que IMAGE_SIZE sea compatible con la red elegida (ej. 224x224 para muchas)
IMAGE_WIDTH_TL = 224 # Tamaño para MobileNetV2, puedes probar otros como 128, 160, 192
IMAGE_HEIGHT_TL = 224
IMAGE_SIZE_TL = (IMAGE_HEIGHT_TL, IMAGE_WIDTH_TL)
IMAGE_CHANNELS_TL = 3 # MobileNetV2 espera 3 canales. Replicaremos una banda si es necesario o usaremos la media.
                     # Ya tenemos X_images con 3 canales (band1, band2, avg)

# Función de preprocesamiento para Transfer Learning
def preprocess_for_tl(image_stack, angle, label):
    # Redimensionar las imágenes
    image_resized = tf.image.resize(image_stack, [IMAGE_HEIGHT_TL, IMAGE_WIDTH_TL])
    # Aplicar preprocesamiento específico de MobileNetV2
    image_preprocessed = mobilenet_preprocess(image_resized)
    return (image_preprocessed, angle), label

# Crear tf.data.Datasets
# BATCH_SIZE ya está definido, podemos reusarlo o definir BATCH_SIZE_TL

# Asegúrate que X_img_train, X_ang_train, y_train etc. están disponibles de la sección 3
train_ds_tl = tf.data.Dataset.from_tensor_slices((X_img_train, X_ang_train, y_train))
train_ds_tl = train_ds_tl.shuffle(buffer_size=len(X_img_train), seed=SEED)
train_ds_tl = train_ds_tl.map(preprocess_for_tl, num_parallel_calls=tf.data.AUTOTUNE)
train_ds_tl = train_ds_tl.batch(BATCH_SIZE)
train_ds_tl = train_ds_tl.prefetch(tf.data.AUTOTUNE)

val_ds_tl = tf.data.Dataset.from_tensor_slices((X_img_val, X_ang_val, y_val))
val_ds_tl = val_ds_tl.map(preprocess_for_tl, num_parallel_calls=tf.data.AUTOTUNE)
val_ds_tl = val_ds_tl.batch(BATCH_SIZE)
val_ds_tl = val_ds_tl.prefetch(tf.data.AUTOTUNE)

test_ds_tl = tf.data.Dataset.from_tensor_slices((X_img_test, X_ang_test, y_test))
test_ds_tl = test_ds_tl.map(preprocess_for_tl, num_parallel_calls=tf.data.AUTOTUNE)
test_ds_tl = test_ds_tl.batch(BATCH_SIZE)
test_ds_tl = test_ds_tl.prefetch(tf.data.AUTOTUNE)

print("Datasets para Transfer Learning creados y preprocesados.")

# Cargar el modelo base MobileNetV2
base_model_tl = MobileNetV2(input_shape=(IMAGE_HEIGHT_TL, IMAGE_WIDTH_TL, IMAGE_CHANNELS_TL),
                            include_top=False, # No incluir el clasificador original de ImageNet
                            weights='imagenet')

# Congelar el modelo base
base_model_tl.trainable = False
print(f"Modelo base MobileNetV2 cargado. Entrenable: {base_model_tl.trainable}")
# --- FIN TU CÓDIGO ---
```

### 6.2. Creación del Nuevo Modelo (Base + Clasificador Propio)

```python
# --- TU CÓDIGO AQUÍ ---
# Define el preprocesamiento específico de la red elegida
def build_transfer_learning_model(base_model, angle_shape, num_classes=1):
    # Entrada para las imágenes (ya preprocesadas para el modelo base)
    image_input_tl = Input(shape=(IMAGE_HEIGHT_TL, IMAGE_WIDTH_TL, IMAGE_CHANNELS_TL), name='image_input_tl')
    # Entrada para el ángulo de incidencia
    angle_input_tl = Input(shape=angle_shape, name='angle_input_tl')

    # Pasar la imagen por el modelo base
    x_img = base_model(image_input_tl, training=False) # training=False para capas congeladas y BatchNorm
    
    # Aplanar o hacer pooling global a la salida del modelo base
    x_img = layers.GlobalAveragePooling2D(name='global_avg_pool')(x_img)
    
    # Concatenar las características de la imagen con el ángulo de incidencia
    concatenated_features_tl = Concatenate(name='concat_tl')([x_img, angle_input_tl])
    
    # Capas densas para clasificación
    x = Dense(128, activation='relu', name='dense_tl_1')(concatenated_features_tl)
    x = Dropout(0.5, name='dropout_tl_1')(x)
    
    # Salida (clasificación binaria)
    output_tl = Dense(num_classes, activation='sigmoid', name='output_tl')(x)
    
    model = Model(inputs=[image_input_tl, angle_input_tl], outputs=output_tl, name="transfer_learning_model")
    return model

# Crear el modelo de Transfer Learning
angle_shape_tl = (1,) 
model_transfer = build_transfer_learning_model(base_model_tl, angle_shape_tl)

model_transfer.summary()
# --- FIN TU CÓDIGO ---
```

### 6.3. Compilación y Entrenamiento (solo del clasificador)

```python
# --- TU CÓDIGO AQUÍ ---
# Compilar el modelo
# LEARNING_RATE y BATCH_SIZE ya deberían estar definidos
# EPOCHS_TRANSFER debería estar definido en la configuración global
optimizer_transfer = tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE)

model_transfer.compile(
    optimizer=optimizer_transfer,
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)
print("Modelo de Transfer Learning compilado (entrenando solo clasificador).")

# Callbacks (early_stopping ya está definido)
# model_checkpoint_path_transfer = os.path.join(MODEL_FOLDER, 'model_transfer_best.keras')
# model_checkpoint_transfer = ModelCheckpoint(
#     filepath=model_checkpoint_path_transfer,
#     monitor='val_auc',
#     save_best_only=True,
#     mode='max',
#     verbose=1
# )

print(f"\nIniciando entrenamiento del modelo de Transfer Learning por {EPOCHS_TRANSFER} épocas...")
# Usamos los tf.data.Dataset que ya tienen las etiquetas
history_transfer = model_transfer.fit(
    train_ds_tl,
    epochs=EPOCHS_TRANSFER,
    validation_data=val_ds_tl,
    callbacks=[early_stopping], # Añadir model_checkpoint_transfer si se usa
    # class_weight se puede usar aquí también si es necesario y si y_train no está one-hot.
    # Pero tf.data.Dataset ya maneja las etiquetas.
    verbose=1
)
# Definir callbacks

# --- FIN TU CÓDIGO ---
```

### 6.4. Evaluación del Modelo de Transfer Learning


```python
# --- TU CÓDIGO AQUÍ ---
print("\n--- Evaluación del Modelo de Transfer Learning (Clasificador Entrenado) ---")

results_transfer = model_transfer.evaluate(test_ds_tl, verbose=0)

print("\nMétricas del modelo de Transfer Learning en el conjunto de Test:")
for name, value in zip(model_transfer.metrics_names, results_transfer):
    print(f"{name}: {value:.4f}")

# Para el reporte de clasificación, necesitamos y_true y y_pred_classes
y_true_test_tl = []
y_pred_probs_tl_list = []

for images_batch, labels_batch in test_ds_tl: # test_ds_tl devuelve ((img, ang), label)
    y_true_test_tl.extend(labels_batch.numpy())
    preds_batch = model_transfer.predict_on_batch(images_batch) # images_batch es (img, ang)
    y_pred_probs_tl_list.append(preds_batch)

y_pred_probs_tl = np.vstack(y_pred_probs_tl_list)
y_pred_classes_tl = (y_pred_probs_tl > 0.5).astype(int).flatten()
y_true_test_tl = np.array(y_true_test_tl)


print("\nReporte de Clasificación (Transfer Learning):")
print(classification_report(y_true_test_tl, y_pred_classes_tl, target_names=target_names_binary))

print("\nMatriz de Confusión (Transfer Learning):")
cm_transfer = confusion_matrix(y_true_test_tl, y_pred_classes_tl)
plt.figure(figsize=(6,5))
sns.heatmap(cm_transfer, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names_binary, yticklabels=target_names_binary)
plt.title('Matriz de Confusión - Transfer Learning')
plt.ylabel('Clase Verdadera')
plt.xlabel('Clase Predicha')
plt.show()

plot_learning_curves(history_transfer, EPOCHS_TRANSFER, title_suffix=' - Transfer Learning (Clasificador)')

metrics_transfer = {
    "model_name": "Transfer Learning (Clasificador)",
    "accuracy": results_transfer[model_transfer.metrics_names.index('compile_metrics')],
    "loss": results_transfer[model_transfer.metrics_names.index('loss')]
}
all_metrics.append(metrics_transfer)

# --- FIN TU CÓDIGO ---
```

### 6.5. Fine-Tuning (Opcional)

```python

# --- TU CÓDIGO AQUÍ ---
print("\n--- Iniciando Fine-Tuning del Modelo de Transfer Learning ---")

# Descongelar parte del modelo base
base_model_tl.trainable = True

# Establecer cuántas capas superiores del base_model descongelar.
# Por ejemplo, las últimas ~50 capas de MobileNetV2 (MobileNetV2 tiene ~155 capas)
# Es una heurística, se puede ajustar.
fine_tune_at_layer = 100 # Descongelar desde la capa 100 en adelante.

for layer in base_model_tl.layers[:fine_tune_at_layer]:
    layer.trainable = False

print(f"Modelo base MobileNetV2 parcialmente descongelado. Capas antes de {fine_tune_at_layer} congeladas.")
for i, layer in enumerate(base_model_tl.layers):
    if i % 20 == 0 or i > fine_tune_at_layer - 5 : # Imprimir algunas para verificar
        print(f"Capa {i}: {layer.name}, Entrenable: {layer.trainable}")


# Re-compilar el modelo con una tasa de aprendizaje muy baja para fine-tuning
LEARNING_RATE_FT = LEARNING_RATE / 10 # ej. 0.0001 si LEARNING_RATE era 0.001
optimizer_ft = tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE_FT)

model_transfer.compile(
    optimizer=optimizer_ft,
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)
print(f"Modelo re-compilado para Fine-Tuning con LR={LEARNING_RATE_FT}.")
model_transfer.summary() # Para ver el estado de las capas entrenables

EPOCHS_FINE_TUNE = 20 
print(f"\nIniciando Fine-Tuning por {EPOCHS_FINE_TUNE} épocas adicionales...")
history_fine_tune = model_transfer.fit(
    train_ds_tl,
    epochs=EPOCHS_TRANSFER + EPOCHS_FINE_TUNE, # Total de épocas
    initial_epoch=history_transfer.epoch[-1] + 1, # Continuar desde la última época
    validation_data=val_ds_tl,
    callbacks=[early_stopping], # Puedes reusar early_stopping o definir uno nuevo, y añadir model_checkpoint_ft
    verbose=1
)

# Evaluación después de Fine-Tuning
print("\n--- Evaluación del Modelo de Transfer Learning (Fine-Tuned) ---")
results_fine_tune = model_transfer.evaluate(test_ds_tl, verbose=0)

print("\nMétricas del modelo Fine-Tuned en el conjunto de Test:")
for name, value in zip(model_transfer.metrics_names, results_fine_tune):
    print(f"{name}: {value:.4f}")

y_true_test_ft_list = []
y_pred_probs_ft_list = []
for images_batch, labels_batch in test_ds_tl:
    y_true_test_ft_list.extend(labels_batch.numpy())
    preds_batch = model_transfer.predict_on_batch(images_batch)
    y_pred_probs_ft_list.append(preds_batch)

y_pred_probs_ft = np.vstack(y_pred_probs_ft_list)
y_pred_classes_ft = (y_pred_probs_ft > 0.5).astype(int).flatten()
y_true_test_ft = np.array(y_true_test_ft_list)

print("\nReporte de Clasificación (Fine-Tuned):")
print(classification_report(y_true_test_ft, y_pred_classes_ft, target_names=target_names_binary))

print("\nMatriz de Confusión (Fine-Tuned):")
cm_fine_tune = confusion_matrix(y_true_test_ft, y_pred_classes_ft)
plt.figure(figsize=(6,5))
sns.heatmap(cm_fine_tune, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names_binary, yticklabels=target_names_binary)
plt.title('Matriz de Confusión - Fine-Tuned')
plt.ylabel('Clase Verdadera')
plt.xlabel('Clase Predicha')
plt.show()

plot_learning_curves(history_fine_tune, EPOCHS_TRANSFER + EPOCHS_FINE_TUNE, title_suffix=' - Transfer Learning (Fine-Tuned)')

metrics_fine_tune = {
    "model_name": "Transfer Learning (Fine-Tuned)",
    "accuracy": results_fine_tune[model_transfer.metrics_names.index('compile_metrics')],
    "loss": results_fine_tune[model_transfer.metrics_names.index('loss')]
}
all_metrics.append(metrics_fine_tune)
# --- FIN TU CÓDIGO ---
```
